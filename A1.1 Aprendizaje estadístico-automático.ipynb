Para comenzar, se importa la librería pandas para extraer las primeras 10 líneas de información dentro de la base de datos.

```python
import pandas as pd
data = pd.read_csv("A1.1 Obesidad.csv")
print(data.head(10))
```


Al analizar estos datos, considero que algunas otras variables que pudieran haber resultado útiles serían las siguientes:
- Porcentaje de grasa corporal (cuantitativa)
- Promedio de horas de sueño (cuantitativa)
- Nivel de actividad física (cualitativa)
- Si se presentan enfermedades asociadas o no (cualitativa)



Si yo fuera el líder del proyecto y quisiera realizar un estudio de inferencia, consideraría que la variable 
más adecuada a considerarse como salida, sería la de NivelDeObesidad, con la cual podemos obtener los promedios 
para cada tipo de nivel de obesidad dentro de la población.

A continuación se muestra la cantidad de personas que se encuentra dentro de cada categoría para la variable NivelDeObesidad:

```python
print(data.NivelDeObesidad.value_counts())
```


Conociendo esto, seleccioné las siguientes variables a graficar:

x = Edad,

y = NivelDeObesidad


```python
%matplotlib inline
import matplotlib.pyplot as plt

plt.scatter(
    x = data.Edad[data.Sexo == "Male"], 
    y = data.NivelDeObesidad[data.Sexo == "Male"], 
    alpha = 0.05
)

plt.show()
```
